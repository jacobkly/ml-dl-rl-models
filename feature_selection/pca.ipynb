```python
# import necessary libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# load the dataset into a pandas DataFrame
df = pd.read_csv('room-temperature.csv')

# drop not needed first column for PCA
df = df.iloc[:, 1:]

# standardize the features
df = StandardScaler().fit_transform(df)
```

```python
# initialize PCA with two components
pca = PCA(n_components=2)

# fit PCA on the standardized data
df_r = pca.fit_transform(df)

# convert into DataFrame
df_r = pd.DataFrame(df_r, columns=['x1', 'x2'])
```

```python
# plot the PCA results with a scatter plot
plt.figure()
plt.scatter(df_r['x1'], df_r['x2'], alpha=.8, color='blue')
plt.xlabel('First Component')
plt.ylabel('Second Component')
plt.title('PCA of Room Temperature Data')
plt.show()
```
